# Voetbal

Wie behoren in het voetbal tot de top 10 aanvallersen de top 10 keepers. Dit zijn de twee hoofdvragen die we in dit document gaan beantwoorden. 

Om deze vragen te gaan beantwoorden gaan we gebruik maken van de European Soccer Database. Hierin vinden we data van:
+25,000 matches
+10,000 players
11 European Countries with their lead championship
Seasons 2008 to 2016
Players and Teams' attributes* sourced from EA Sports' FIFA video game series, including the weekly updates
Team line up with squad formation (X, Y coordinates)
Betting odds from up to 10 providers
Detailed match events (goal types, possession, corner, cross, fouls, cards etc.) for +10,000 matches

Laten we beginnen met de vraag: 

## **1. Wie zijn de beste 10 aanvallers?** 

Aanvallers zijn de spelers van een voetbalteam die het dichtst bij het doel van de tegenpartij spelen en daarom het meest verantwoordelijk zijn voor het maken van doelpunten. Door hun geavanceerde positie en beperkte verdedigende verantwoordelijkheden maken aanvallers normaal gesproken meer doelpunten namens hun team dan andere spelers. 

De European Soccer Database bevat meerdere files, maar we willen allereerst gaan kijken naar de gegevens van de spelers. Deze gegevens zijn verdeeld over twee bestanden. In het ene bestand staan enkele demografische gegevens als naam van de speler, gewicht, lengte etc. In het andere bestand staan de gegevens van verschillende onderdelen zoals overall rating, dribbelen, koppen etc. 

Om te beginnen zal ik beide documenten samenvoegen zodat we meteen alle gegevens op één plek hebben staan en meteen kunnen zien over welke spelers het gaat. 

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("database.sqlite")
df = pd.read_sql_query("SELECT * from Player_Attributes", con)
df1 = pd.read_sql_query("SELECT * from Player", con)
con.close()

# Combineer df1 en aanvallers, met als key variable player_fifa_api_id zodat we de namen van de voetballers kunnen zien
df_namen = pd.merge(df, df1, on="player_fifa_api_id")
df_namen
```

Wat opvalt is dat spelers vaker in de lijst voorkomen. Dit komt omdat er op verschillende datums gegevens verzameld zijn voor die bepaalde speler. Om ervoor te zorgen dat een speler maar één keer in de lijst voorkomt kunnen we verschillende dingen doen: 

- We kunnen een bepaalde datum selecteren
Als we een bepaalde datum gaan selecteren bestaat de kans dat niet alle spelers in de lijst voorkomen. Daarnaast kunnen we uit bovenstaande lijst al opmaken dat het selecteren op jaar ook lastig wordt, gezien er meerdere keren per jaar gegevens verzameld zijn. Selecteren op datum heeft daarom niet mijn voorkeur. 

- We kunnen de hoogste scores van een speler op alle onderdelen selecteren
We zouden ervoor kunnen kiezen om van een speler alleen de hoogste scores van een variabele te selecteren. Hierdoor zou je een lijst kunnen overhouden met scores die op verschillende momenten gemeten zijn. In mijn ogen zou dat voor een vertekend beeld kunnen zorgen.  

- Alle gegevens van een speler samenvoegen en het gemiddeld resultaat selecteren
We zouden alle gegevens van een bepaalde speler bij elkaar kunnen verzamelen en vervolgens een gemiddelde score voor elke variabele berekenen. We moeten dan wel echt benadrukken dat het over een top 10 gaat van een bepaalde periode, 2008 tot en met 2016. Voor- of achteruitgang van een speler zal namelijk de totaal score van een speler beïnvloeden. 

Deze laatste optie is in mijn ogen de meest geschikte omdat ik op deze manier alle spelers en alle gegevens kan meenemen. Ik ga nu een nieuwe lijst aanmaken die per speler de gemiddelde resultaten van een bepaalde variabele zal weergeven.

```python
# We willen de gemiddelde resulten van elke speler gebruiken
df_speler = df_namen.groupby('player_name').mean()
df_speler
```

Nu we een complete lijst hebben van alle spelers en alle variabelen wil ik gaan inzoomen op de skills die belangrijk zijn voor een aanvaller. Deze skills ofwel variabelen zal ik uit de volgende lijst selecteren:

```python
for col in df_speler.columns:
    print(col)
```

De variabelen die ik zal meenemen om een lijst te maken van de top 10 aanvallers zijn:
- finishing
- sprint_speed
- dribbling
- ball_control
- heading_accuracy

Een goede aanvaller is namelijk snel, kan goed dribbelenen, heeft controle over de bal, is in staat om gericht te koppen en kan een goal scoren. Met name finishing ofwel het scoren van een goal en sprint_speed ofwel snelheid vind ik kwaliteiten die een goede aanvaller zeker moet hebben. Omdat ik deze twee variabelen zo belangrijk vindt zal ik deze zwaarder laten meewegen tijdens het opstellen van een nieuwe rating lijst. 

```python
# Een nieuwe lijst maken met de gekozen variabelen, waarbij de variabelen finishen en sprint_speed zwaarder meewegen dan de andere variabelen 
df_speler['goal'] = df_speler['finishing']*0.3
df_speler['snelheid'] =df_speler['sprint_speed']*0.3
df_speler['dribbelen'] = df_speler['dribbling']*0.2
df_speler['controle'] = df_speler['ball_control']*0.1
df_speler['koppen'] = df_speler['heading_accuracy']*0.1

df_aanvaller = df_speler[['player_fifa_api_id', 'goal', 'snelheid', 'dribbelen', 'controle', 'koppen']].copy()

# De variabelen goal, snelheid, dribbelen, koppen en versnellen per id bij elkaar optellen 
aanvaller_totaal = df_aanvaller['goal'] + df_aanvaller['snelheid'] + df_aanvaller['dribbelen'] + df_aanvaller['controle'] + df_aanvaller['koppen']
df_aanvaller['totaal'] = aanvaller_totaal
aanvallers = df_aanvaller.sort_values(by = 'totaal', ascending = False)
aanvallers.head(10)
```

We hebben nu een nieuwe lijst gemaakt met alleen maar variabelen die belangrijk zijn voor een goede aanvaller. Alle variabelen zijn bij elkaar opgeteld en aan de hand van deze totaal score kunnen we een 'Aanvallers top 10' vaststellen. 

De top 10 aanvallers zijn: 
1. Christiano Ronaldo
2. Lionel Messi
3. Sergio Aguero
4. Zlatan Ibrahimovic
5. Neymar
6. Arjan Robben
7. Karim Benzema
8. Samuel Eto'o
9. Nilmar
10. Pierre-Emerick Aubameyang

### 1.1 Excelleert de beste aanvaller op alle onderdelen?

Nu we weten wie de top 10 aanvallers van Europa zijn tussen 2008 en 2016 ben ik benieuwd of de beste aanvaller ook op alle punten de hoogste score behaald. Met andere woorden: 'Scoort Christiano Ronaldo het hoogst op snelheid, dribbelen, balcontrole, koppen en het scoren van een goal?'. We kunnen dit uiteraard uit de lijst opmaken, maar het is makkelijker om dit aan de hand van een figuur te bekijken.

```python
# Reset index
aanvallers_top10 = aanvallers.reset_index()

# figuur grootte
plt.rcParams["figure.figsize"] = (10, 5)

# creating plot 1
Aanvallers = aanvallers_top10['player_name'].head(10)
Goal = aanvallers_top10['goal'].head(10)

plt.title("Goal")
clrs = ['grey' if (x < max(Goal)) else 'green' for x in Goal]
plt.barh(Aanvallers, Goal, color=clrs,)
plt.show()

# creating plot 2
Aanvallers = aanvallers_top10['player_name'].head(10)
Snelheid = aanvallers_top10['snelheid'].head(10)

plt.title("Snelheid")
clrs = ['grey' if (x < max(Snelheid)) else 'green' for x in Snelheid]
plt.barh(Aanvallers, Snelheid, color=clrs,)
plt.show()

# plot 3
Aanvallers = aanvallers_top10['player_name'].head(10)
Dribbelen = aanvallers_top10['dribbelen'].head(10)

plt.title("Dribbelen")
clrs = ['grey' if (x < max(Dribbelen)) else 'green' for x in Dribbelen]
plt.barh(Aanvallers, Dribbelen, color=clrs,)
plt.show()

#plot 4
Aanvallers = aanvallers_top10['player_name'].head(10)
Controle = aanvallers_top10['controle'].head(10)

plt.title("Controle")
clrs = ['grey' if (x < max(Controle)) else 'green' for x in Controle]
plt.barh(Aanvallers, Controle, color=clrs,)
plt.show()

#plot 5
Aanvallers = aanvallers_top10['player_name'].head(10)
Koppen = aanvallers_top10['goal'].head(10)

plt.title("Koppen")
clrs = ['grey' if (x < max(Koppen)) else 'green' for x in Koppen]
plt.barh(Aanvallers, Koppen, color=clrs,)
plt.show()
```

Uit bovenstaande figuren kunnen we opmaken dat de plek voor de beste aanvaller met name ging tussen Lionel Messi en Christiano Ronaldo. Lionel Messi scoort op alle onderdelen, behalve snelheid beter dan Christiano Ronaldo. Het feit dat we de variabelen snelheid en goal zwaarder hebben laten meewegen hebben er uiteindelijk voor gezorgd dat Christiano Ronaldo in deze lijst als beste aanvaller naar voren komt. 

### **1.2 Conclusie** 
Aan de hand van de variabelen: het scoren van een goal, snelheid (sprint), dribbelen, controle over de bal en koppen, blijkt dat (middels de gegevens van de European Soccer Database), Christiano Ronaldo de beste aanvaller is in de periode 2008 tot en met 2016.

## 2. Wie is de beste keeper?

Wie is de beste keeper? We zullen bij het beantwoorden van deze vraag wederom gebruik maken van de European Soccer Database. In het Player_Attributes bestand staan ook gegevens over keepers. Zo zijn er gegevens verzameld over de volgende keepers skills:
- gk_diving 
- gk_handling
- gk_kicking
- gk_positioning
- gk_reflexes

Hoewel alle variabelen belangrijk zijn voor een goede keeper, vind ik gk_handling het meest belangrijke. Goal keeper handling is een spelerskenmerk in FIFA dat bepaalt in hoeverre een speler in staat is om de bal te kunnen hanteren als keeper. In principe dus de vaarigheid van de keeper. Omdat ik deze variabele zo belangrijk vind zal ik bij het opstellen van een 'Keepers top 10' deze variabele (gk_handling) zwaarder laten meewegen. 

```python
# Een nieuwe lijst maken met de keepers variabelen
df_speler['gk_duiken'] = df_speler['gk_diving']*0.1
df_speler['gk_vaardigheden'] =df_speler['gk_handling']*0.6
df_speler['gk_schieten'] = df_speler['gk_kicking']*0.1
df_speler['gk_positie'] = df_speler['gk_positioning']*0.1
df_speler['gk_reflexen'] = df_speler['gk_reflexes']*0.1

df_keeper = df_speler[['player_fifa_api_id', 'gk_duiken', 'gk_vaardigheden', 'gk_schieten', 'gk_positie', 'gk_reflexen']].copy()

#de keepers variabelen per player_fifa_api_id bij elkaar optellen
keeper_totaal = df_keeper['gk_duiken'] + df_keeper['gk_vaardigheden'] + df_keeper['gk_schieten'] + df_keeper['gk_positie'] + df_keeper['gk_reflexen']
df_keeper['totaal'] = keeper_totaal
keepers = df_keeper.sort_values(by = 'totaal', ascending = False)
keepers.head(10)
```

### 2.1 Excelleert de beste keeper op alle onderdelen?

Scoort de beste keeper ook het hoogst op alle onderdelen of zijn er keepers die op sommige onderdelen beter scoren?


```python
# Reset index
keepers_top10 = keepers.reset_index()

# creating plot 1
Keepers = keepers_top10['player_name'].head(10)
Duiken = keepers_top10['gk_duiken'].head(10)

plt.title("Duiken")
clrs = ['grey' if (x < max(Duiken)) else 'green' for x in Duiken]
plt.barh(Keepers, Duiken, color=clrs,)
plt.show()

# creating plot 2
Keepers = keepers_top10['player_name'].head(10)
Vaardigheden = keepers_top10['gk_vaardigheden'].head(10)

plt.title("Vaardigheden")
clrs = ['grey' if (x < max(Vaardigheden)) else 'green' for x in Vaardigheden]
plt.barh(Keepers, Vaardigheden, color=clrs,)
plt.show()

# creating plot 3
Keepers = keepers_top10['player_name'].head(10)
Schieten = keepers_top10['gk_schieten'].head(10)

plt.title("Schieten")
clrs = ['grey' if (x < max(Schieten)) else 'green' for x in Schieten]
plt.barh(Keepers, Schieten, color=clrs,)
plt.show()

# creating plot 4
Keepers = keepers_top10['player_name'].head(10)
Positie = keepers_top10['gk_positie'].head(10)

plt.title("Positie")
clrs = ['grey' if (x < max(Positie)) else 'green' for x in Positie]
plt.barh(Keepers, Positie, color=clrs,)
plt.show()

# creating plot 5
Keepers = keepers_top10['player_name'].head(10)
Reflexen = keepers_top10['gk_reflexen'].head(10)

plt.title("Reflexen")
clrs = ['grey' if (x < max(Reflexen)) else 'green' for x in Reflexen]
plt.barh(Keepers, Reflexen, color=clrs,)
plt.show()
```

Uit bovenstaande figuren kunnen we opmaken dat Edwin van der Sar op 2 van de 5 onderdelen het hoogste scoort, dit zijn de onderdelen vaardigheden en positie. Bij het onderdeel schieten behoort Edwin van der Sar tot de hoogste scores. Echter op de onderdelen reflexen en duiken behoort hij niet bij de hoogste scores, sterker nog, hij behoort op die onderdelen niet eens bij de top 5. De reden waarom hij toch als beste keeper van onze dataset naar voren komt is omdat ik ervoor gekozen heb het onderdeel vaardigheden ofwel gk_handling een stuk zwaarder te laten meewegen dan de andere onderdelen. Had ik hier niet voor gekozen bestaat er een grote kans dat Edwin van der Sar niet als beste keeper naar voren was gekomen. 

### 2.2 Conclusie

Op basis van alle gegevens van de European Soccer Database kunnen we vaststellen dat Edwin van der Sar de beste keeper is aan de hand van de keeper skills: duiken, keepers vaardigheden, schieten, positie en reflexen.
